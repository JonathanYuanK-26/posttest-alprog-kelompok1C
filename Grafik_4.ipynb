# Grafik 4 - Boxplot price

Jalankan semua sel dari atas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Kelas C_Housing.csv")
df.head()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 6))
ax = axes[1, 1]
ax.boxplot(df["price"])
ax.set_title("Distribusi Price dan Pencilan")
ax.set_ylabel("Price")
ax.set_xticklabels(["price"])
axes[0, 0].axis("off")
axes[0, 1].axis("off")
axes[1, 0].axis("off")
plt.tight_layout()
```
