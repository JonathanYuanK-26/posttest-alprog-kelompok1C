# Grafik 4 - Boxplot price

Jalankan semua sel dari atas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

df = pd.read_csv("Kelas C_Housing.csv", sep=",")
df.head()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 6))
ax = axes[1, 1]
sns.boxplot(data=df, y="price", ax=ax)
ax.set_title("Distribusi Price dan Pencilan")
ax.set_xlabel("")
ax.set_ylabel("Price")
axes[0, 0].axis("off")
axes[0, 1].axis("off")
axes[1, 0].axis("off")
fig.tight_layout()
```
