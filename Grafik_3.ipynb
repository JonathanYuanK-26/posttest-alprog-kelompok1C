# Grafik 3 - Korelasi area vs price

Jalankan semua sel dari atas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Kelas C_Housing.csv")
df.head()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 6))
ax = axes[1, 0]
ax.scatter(df["area"], df["price"])
ax.set_title("Area vs Price")
ax.set_xlabel("Area")
ax.set_ylabel("Price")
axes[0, 0].axis("off")
axes[0, 1].axis("off")
axes[1, 1].axis("off")
plt.tight_layout()
```
