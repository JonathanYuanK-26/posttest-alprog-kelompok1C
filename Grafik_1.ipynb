# Grafik 1 - Rata-rata price per furnishingstatus

Jalankan semua sel dari atas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

df = pd.read_csv("Kelas C_Housing.csv", sep=",")
df.head()
```

```python
price_by_furniture = (
    df.groupby("furnishingstatus", as_index=False)["price"]
    .mean()
    .sort_values("price", ascending=False)
)

fig, axes = plt.subplots(2, 2, figsize=(8, 6))
ax = axes[0, 0]
sns.barplot(data=price_by_furniture, x="furnishingstatus", y="price", ax=ax)
ax.set_title("Rata-rata Harga per Furnishing")
ax.set_xlabel("Furnishing Status")
ax.set_ylabel("Price")
axes[0, 1].axis("off")
axes[1, 0].axis("off")
axes[1, 1].axis("off")
fig.tight_layout()
```
