# Grafik 1 - Rata-rata price per furnishingstatus

Jalankan semua sel dari atas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Kelas C_Housing.csv")
df.head()
```

```python
price_by_furniture = df.groupby("furnishingstatus")["price"].mean().sort_values(ascending=False)

fig, axes = plt.subplots(2, 2, figsize=(8, 6))
ax = axes[0, 0]
ax.bar(price_by_furniture.index, price_by_furniture.values)
ax.set_title("Rata-rata Harga per Furnishing")
ax.set_xlabel("Furnishing Status")
ax.set_ylabel("Price")
axes[0, 1].axis("off")
axes[1, 0].axis("off")
axes[1, 1].axis("off")
plt.tight_layout()
```
