# Grafik 2 - Filter mainroad/AC/parking

Jalankan semua sel dari atas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

df = pd.read_csv("Kelas C_Housing.csv", sep=",")
df.head()
```

```python
filtered = df[
    (df["mainroad"] == "yes")
    & (df["airconditioning"] == "yes")
    & (df["parking"] >= 2)
]

filtered_mean = filtered["price"].mean()
filtered_bar = pd.DataFrame(
    {"kategori": ["mainroad=yes, AC=yes, parking>=2"], "mean_price": [filtered_mean]}
)

fig, axes = plt.subplots(2, 2, figsize=(8, 6))
ax = axes[0, 1]
sns.barplot(data=filtered_bar, x="kategori", y="mean_price", ax=ax)
ax.set_title("Rata-rata Harga (Filter)")
ax.set_xlabel("")
ax.set_ylabel("Price")
ax.set_xticklabels(ax.get_xticklabels(), rotation=20, ha="right")
axes[0, 0].axis("off")
axes[1, 0].axis("off")
axes[1, 1].axis("off")
fig.tight_layout()
```
